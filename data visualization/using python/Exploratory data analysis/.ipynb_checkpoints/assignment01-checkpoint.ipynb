```python
# chybƒõj√≠c√≠ hodnoty v datasetech p≈ôepi≈°u na NaN
intakes = intakes.replace('Unknown', np.nan)
# intakes = intakes.replace('Other', np.nan)

outcomes = outcomes.replace('Unknown', np.nan)
# outcomes = outcomes.replace('Other', np.nan)
```

### Zkonvertuju kategorick√© p≈ô√≠znaky na typ *category*

```python
for col in ['Name', 'Found Location', 'Intake Type', 'Intake Condition', 'Animal Type', 'Sex upon Intake', 'Breed', 'Color']:
    intakes[col] = intakes[col].astype('category')
```

```python
for col in ['Name', 'Outcome Type', 'Outcome Subtype', 'Animal Type', 'Sex upon Outcome', 'Breed', 'Color']:
    outcomes[col] = outcomes[col].astype('category')
```
